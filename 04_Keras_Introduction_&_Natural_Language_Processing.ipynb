**tf.keras.layers.Dense**

$$y = f(Wx +b)$$  
* x와 b는 input Vector, w는 Weight Matrix
* 신경망 구조의 가장 기본적인 형태 
* 단일 Layer (Hidden Layer가 없는) 의 형태로 이뤄진 신경망 Diagram 
* 주요 인자
    * units : 출력값의 크기, integer/long 형태
    * activation : 활성화 함수
    * use_bias : 편향의 사용여부 True/False
    * kernel_initializer : 가중치 초기화 함수 
    * bias_initializer : 편향 초기화 함수
    * kernel_regularizer : 가중치 정규화 방법
    * bias_regularizer : 편향 정규화 방법
    * activity_regularizer : 출력 값 정규화 방법
    * kernel_constraint : Optimizer에 의해 업데이트 된 이후에 가중치에 적용되는 부가적 제약함수 
    * bias_constraint :Optimizer에 의해 업데이터 된 이후에 편향에 적용되는 부가적 제약함수 

**tf.keras.layers.Dropout**

* 신경망에서 Overfitting(과적합)문제를 Regularization(정규화)방식으로 해결할때 사용되는 방법
* 학습할때만 적용되며, 특정 노드들의 값을 0으로 변환 
* 주요 인자 
    * rate : 드럽아웃을 적용할 확률 지정 (설정된 확률에 의해 노드의 해당 확률(비율)에 해당하는 값이 0으로 변환)
    * noise_shape : 정수형의 1D tensor값을 받아 특정 채널에 Dropout
    * seed : 임의의 선택된 Seed 값으로 dropout
   

**tf.keras.layers.Conv1D**

## 단어표현 (Word Representation)

* 단어 표현 : 모든 자연어 처리의 기본 바탕, 기계가 단어를 어떻게 처리하여 표현하는지 
* 컴퓨터의 이해 방식 : 이진법 / 유니코드 / 아스키코드
* 자연어는 기본적으로 유니코드를 이용해 컴퓨터가 알아들을 수 있는 이진법으로 변환되어 입력된다. 
* 언어적 특성을 반영하여 단어를 수치화. 주로 **Vector**를 이용해 단어를 수치화 한다. 
* 단어표현은 **word embedding**또는 **word vector**로도 통용된다. 


* One Hot Encoding : 단어를 표현하는 가장 기본적인 방식. 단어를 1 또는 0으로 표현한다. 
    - 일정 단어의 index를 지정한 뒤, 해당 단어를 표현하는 기법 
    - 문제점 : 수많은 단어를 표현하기에 index가 너무 많이 들어간다 (비효율적) / 단어의 특성이나 의미를 나타내지는 않음 


* Distributed Hypothesis : 같은 문맥의 단어, 비슷한 위치의 단어가 나올 확률을 계산(유사도 계산)을 통해 의미 파악 
    - Count Base : 문장 내 단어등장의 횟수를 이용 
    - Predictive : 알고리즘 Modeling을 통한 문장 내 단어 예측 

### Count Base 방법 

* 문장 내 단어가 얼마나 동시에 발생하는지, 단어 출현의 빈도수 계산 
* 단어동시출현횟수 (Co occurrence) Matrix를 생성하고, 그 Matrix를 변형하는 방식
* 분석이 빠르며, 단어가 잘 표현되는 효율적 방식 중 하나 
* Count Base의 여러 방식 
    - 특이값 분해 (Singular Value Decomposition, SVD)
    - 잠재의미분석 (Latent Semantic Analysis, LSA)
    - Hypoerspace Analogue to Language (HAL) 하이퍼 아날로그 랭귀지 
    - Hellinger Principal Component Analysis (헬링거 PCA)

### Predictive 방법

* 신경망 알고리즘을 이용하여, 특정 문맥에서 어떤 단어가 등장할지 예측하면서 단어를 Vector로 만드는 방식
* 일반적으로 Count Base방식보다 성능이 좋다. 
* Predictive 여러 방식
    - Word2vec
    - NNLM (Neural Network Language Model)
    - RNNLM (Recurrent Neural Network Language Model)
   

**Word2vec**

* 단어들을 예측하면서 Word Vector를 계속 학습
* CBOW(Continuous Bag of Words) : 문맥안의 주변 단어들을 통해 특정 단어를 예측하는 방법
   - 종찬이는 아무생각 없이 유튜브를 시작했다. 
   - 종찬이는 아무생각 없이 ------를 시작했다. 
   - 학습 방식 : 
       1. 각 주변 단어들을 One hot Vector로 만들어 입력값으로 사용한다 (입력층 벡터) 
       2. 가중치 행렬(Weight Matrix)를 각 One hot Vector에서 곱해서 n-dimansion Vector를 만든다 (N-D Hidden Layer)
       3. 만들어진 N-d Vector를 모두 더한 후 개수로 나눠 평균 N-d Vector를 만든다 (Label Vector)
       4. N-d Vector에서 다시 Weight Matrix를 곱해서 One-hot Vector와 같은 차원의 Vector를 생성한다. 
       5. 만들어진 Vector를 예측하려고 하는 단어의 One hot Vector와 비교하여 학습한다.
       
   
* Skip-Gram Model : 어떤 단어를 이용해 문맥안 주변 단어들을 예측하는 방법
    - 종찬이는 아무생각 없이 유튜브를 시작했다.
    - ---------------------- 유튜브를 ---------
    - 학습 방식 : 
        1. 하나의 단어를 One Hot Vector로 만들어서 입력값으로 사용한다. (Input Layer Vector) 
        2. Weight Matrix를 One hot Vector에 곱해서 N-d Vector를 만든다. (N-d Hidden Layer)
        3. n-d Vector에 다시 Weight Matrix를 곱해서 One hot Vector와 같은 차원의 Vector로 만든다 (Label Vector)
        4. 만들어진 Vector를 실제 예측하려는 주변 단어들 각각의 One hot Vectorㅇ롸 비교해서 학습한다. 
  
  
* CBOW는 입력값으로 여러개의 단어를 사용, 학습을 위해 하나의 단어와 비교한다.
* Skip-Gram Model은 입력값이 하나의 단어, 학습을 위해 주변의 여러 단어와 비교한다. 
* Count Base 방식보다 단어 간 유사도를 더욱 잘 측정하며, 단어들의 복잡한 특징까지도 도출. 서로 유의미한 관계를 측정할 수 있음 
* 일반적인 경우엔, Skip-Gram Model을 많이 사용한다.

```python
X = df2[['text']]
Y = df2[['label']]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state= 1234)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

**형태소 분석 및 워드클라우드**

```python
okt = Okt()
```

**데이터 전처리 (토크나이징 & 패딩)**

```python
# 불용어 및 특수문자 제거 (직접 작업하면서 제거해야함)
def text_prepro(i, remove_stopword= False):
    okt = Okt()
    review_text = re.sub("[!1(+.),.@@\n2345']?","",i)
    word_text = okt.morphs(review_text,stem=True)

    if remove_stopword:
        word_text = [token for token in word_text if not token in set(['가','은','는','이', '수','인','함','을','를','에','적','만','하','아','것','들','의','있','나','되','등','한','극'])]
    
    return word_text
```

```python
okt = Okt()
clean_train_review = []
clean_test_review = [] 

for i in X_train['text']:
    if type(i) == str:
        clean_train_review.append(text_prepro(i, remove_stopword=True))
    else:
        clean_train_review.append([])
        
for i in X_test['text']:
    if type(i) == str:
        clean_test_review.append(text_prepro(i, remove_stopword=True))
    else:
        clean_test_review.append([])
```

```python
# 인덱스 벡터로 변환 뒤 ,Padding 처리 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_train_review)

train_sequence = tokenizer.texts_to_sequences(clean_train_review)
test_sequence = tokenizer.texts_to_sequences(clean_test_review)

word_vector =tokenizer.word_index
```

```python
#문장 최대길이 설정 (리뷰의 평균 길이)
Max_length = 70
```

```python
train_input = pad_sequences(train_sequence, maxlen=Max_length, padding='post')
train_labels = Y_train['label'].values

test_input = pad_sequences(test_sequence, maxlen=Max_length, padding='post')
test_labels = Y_test['label'].values
```

**모델링 실시**

```python
x = train_input
y = train_labels
```

```python
model = Sequential()
model.add(layers.Dense(100,activation='relu'))
model.add(layers.Dense(60,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
history = model.fit(x, y, epochs=50, batch_size=128, validation_split=0.3)
```

```python
history_table = pd.DataFrame(history.history)
val_set = history_table[['val_loss','val_acc']].rename(columns= {'val_loss':'loss','val_acc':'acc'}).reset_index()
train_set = history_table[['loss','acc']].reset_index()
train_set['label'] = 'train'
val_set['label'] = 'validation'
result = pd.concat([train_set,val_set])
result.head(10)
```
